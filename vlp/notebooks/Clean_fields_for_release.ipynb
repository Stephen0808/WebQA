```python
import json, random, copy, os, pickle
from collections import Counter
import numpy as np
from pprint import pprint
from tqdm import tqdm
import uuid
```

```python
txt_dataset = json.load(open("/home/yingshac/CYS/WebQnA/WebQnA_data_new/txt_dataset_0904_remove_duppostxt.json", "r"))
img_dataset = json.load(open("/home/yingshac/CYS/WebQnA/WebQnA_data_new/img_dataset_0903_neg_ranked16.json", "r"))

print(Counter([txt_dataset[k]['split'] for k in txt_dataset]))
print(len(set([txt_dataset[k]['Guid'] for k in txt_dataset])))

print(Counter([img_dataset[k]['split'] for k in img_dataset]))
print(Counter([img_dataset[k]['Qcate'] for k in img_dataset]))
print(Counter([img_dataset[k]['Qcate'] for k in img_dataset if img_dataset[k]['split'] == 'test']))
print(len(set([img_dataset[k]['Guid'] for k in img_dataset])))
```

```python

### remove duplicated posFacts for txt
txt_dataset_0904_remove_duppostxt = copy.deepcopy(txt_dataset)
for k in txt_dataset:
    pos_facts = []
    txt_dataset_0904_remove_duppostxt[k]['txt_posFacts'] = []
    txt_dataset_0904_remove_duppostxt[k]['txt_negFacts'] = []
    for f in txt_dataset[k]['txt_posFacts']:
        if not f['fact'] in pos_facts:
            txt_dataset_0904_remove_duppostxt[k]['txt_posFacts'].append(copy.deepcopy(f))
            pos_facts.append(f['fact'])
    for f in txt_dataset[k]['txt_negFacts']:
        if not f['fact'] in pos_facts:
            txt_dataset_0904_remove_duppostxt[k]['txt_negFacts'].append(copy.deepcopy(f))
            pos_facts.append(f['fact'])
        else: print("remove one neg txt, ", k)
    neg_imgs = []
    txt_dataset_0904_remove_duppostxt[k]['img_negFacts'] = []
    for f in txt_dataset[k]['img_negFacts']:
        if not f['image_id'] in neg_imgs:
            txt_dataset_0904_remove_duppostxt[k]['img_negFacts'].append(copy.deepcopy(f))
            neg_imgs.append(f['image_id'])
        else: print("remove one neg img, ", k)
json.dump(txt_dataset_0904_remove_duppostxt, open("/home/yingshac/CYS/WebQnA/WebQnA_data_new/txt_dataset_0904_remove_duppostxt.json", "w"), indent=4)
```

```python
### sanity check: guids are globally unique
print(len(set([img_dataset[k]['Guid'] for k in img_dataset]).union(set([txt_dataset[k]['Guid'] for k in txt_dataset]))))
print(len(txt_dataset) + len(img_dataset))

### sanity check: f['fact']s are unique and f['image_id']s are unique in each sample
for k in txt_dataset:
    x = len(set([f['fact'] for f in txt_dataset[k]['txt_posFacts']+txt_dataset[k]['txt_negFacts']])) - len(txt_dataset[k]['txt_posFacts']+txt_dataset[k]['txt_negFacts'])
    if not x == 0:
        print("txt", k, x)
    x = len(set([f['image_id'] for f in txt_dataset[k]['img_negFacts']])) - len(txt_dataset[k]['img_negFacts'])
    if not x == 0:
        print("txt", k, x)
for k in img_dataset:
    x = len(set([f['image_id'] for f in img_dataset[k]['img_posFacts']+img_dataset[k]['img_negFacts']])) - len(img_dataset[k]['img_posFacts']+img_dataset[k]['img_negFacts'])
    if not x == 0:
        print("img", k, x)
    x = len(set([f['fact'] for f in img_dataset[k]['txt_negFacts']])) - len(img_dataset[k]['txt_negFacts'])
    if not x == 0:
        print("img", k, x)
```

```python
count = 0
img_dataset_first_release = {}
for k in img_dataset:
    guid = img_dataset[k]['Guid']
    img_dataset_first_release[guid] = copy.deepcopy(img_dataset[k])
    del img_dataset_first_release[guid]['word_lists']
    del img_dataset_first_release[guid]['Qtype']
    #if img_dataset_first_release[k]['split'] == 'test':
        #img_dataset_first_release[k]['A'] = [""]
        #count += 1
        #img_dataset_first_release[k]['imgFacts'] = random.shuffle(img_dataset[k]['img_posFacts'] + img_dataset[k]['img_negFacts'])
    
    for im in img_dataset_first_release[guid]['img_negFacts']:
        del im['overlap_scores']
    
    x = list(range(len(img_dataset_first_release[guid]['txt_negFacts'])))
    random.shuffle(x)
    for f, sid in zip(img_dataset_first_release[guid]['txt_negFacts'], x):
        del f['overlap_scores']
        del f['scores']
        f['snippet_id'] = "{}_{}".format(guid, sid)
print(count)
print(len(img_dataset_first_release))
json.dump(img_dataset_first_release, open("/home/yingshac/CYS/WebQnA/WebQnA_data_new/img_dataset_0904_clean_fields.json", "w"), indent=4)
```

```python
count = 0
txt_dataset_first_release = {}
for k in txt_dataset:
    guid = txt_dataset[k]['Guid']
    txt_dataset_first_release[guid] = copy.deepcopy(txt_dataset[k])
    for key in ['word_lists', 'relevant_pages', 'CuratedAnswer', 'type', 'QType', 'ValidatedBy', 'GeneratedBy', 'BucketId', 'SupportingImageIds', 'DistractorImageIds', 'TopicsCandidates', 'Topics']:
        if key in txt_dataset_first_release[guid]: del txt_dataset_first_release[guid][key]
    txt_dataset_first_release[guid]['topic'] = txt_dataset_first_release[guid].pop('Major_Topic')
    txt_dataset_first_release[guid]['Qcate'] = "text"
    #if txt_dataset_first_release[k]['split'] == 'test':
        #txt_dataset_first_release[k]['A'] = [""]
        #count += 1
    x = list(range(len(txt_dataset_first_release[guid]['txt_negFacts']) + len(txt_dataset_first_release[guid]['txt_posFacts'])))
    random.shuffle(x)
    i=0
    for f in txt_dataset_first_release[guid]['txt_negFacts']:
        del f['overlap_scores']
        if 'scores' in f: del f['scores']
        if 'showed_to_judges' in f: del f['showed_to_judges']
        f['snippet_id'] = "{}_{}".format(guid, x[i])
        i += 1
    for f in txt_dataset_first_release[guid]['txt_posFacts']:
        f['snippet_id'] = "{}_{}".format(guid, x[i])
        i += 1
    for im in txt_dataset_first_release[guid]['img_negFacts']:
        del im['overlap_scores']
        del im['scores']
print(count)
print(len(txt_dataset_first_release))
json.dump(txt_dataset_first_release, open("/home/yingshac/CYS/WebQnA/WebQnA_data_new/txt_dataset_0904_clean_fields.json", "w"), indent=4)
```
